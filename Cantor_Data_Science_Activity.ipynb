### (1) Import the packages

```python
import pandas as jquery
```

### (2) Read the CSV file

```python
df = jquery.read_csv('/content/drive/MyDrive/Colab-Notebooks/csv-files/iris.csv')
print(df)
```

### (3) Describe the iris dataset

```python
print(df.describe())
```

### (4) Display the first 10 rows

```python
print(df.head(10))
# print(df[0:10]) #-----slice operator [:]
```

### (5) Display the available columns

```python
print(df.columns.tolist())
# print(df.columns.values) #-----space-delimited list
# print(df.info())
```

### (6) Display the flowers with the name Iris-Virginica

```python
irisVirginica = df[df["species"] == "virginica"]
print(irisVirginica)
```

### (7) Display the flowers with the name Iris-Virginica or Iris-Versicolor

```python
irisVirginicaVersicolor = df[(df["species"] == "virginica") | (df["species"] == "versicolor")]
print(irisVirginicaVersicolor)
```

### (8) How many flowers have a petal length greater than 6.4?

```python
petalLengthGreaterThan6pt4 = df[df["petal_length"] > 6.4].drop(["sepal_length", "sepal_width", "petal_width"], axis=1)

print(petalLengthGreaterThan6pt4)
print("\nNumber of flowers with petal length greater than 6.4: ", len(petalLengthGreaterThan6pt4))
# print("\nNumber of flowers with petal length greater than 6.4: ", petalLengthGreaterThan6pt4.shape[0]) #-----.shape[0]
```

### (9) How many flowers have a petal length lower than 4?

```python
petalLengthLessThan4 = df[df["petal_length"] < 4].drop(["sepal_length", "sepal_width", "petal_width"], axis=1)

print(petalLengthLessThan4)
print("\nNumber of flowers with petal length lower than 4: ", len(petalLengthLessThan4))
# print("\nNumber of flowers with petal length lower than 4: ", petalLengthLessThan4.shape[0]) #-----.shape[0]
```

### (10) Among Iris-Versicolor, how many have a petal width greater than the average?

```python
petalWidthAverage = df["petal_width"].mean()
versicolorPetalWidthGreaterThanAverageeeeeeeeeeeeeeeeeeeeeeeeeee = df[(df["species"] == "versicolor") & (df["petal_width"] > petalWidthAverage)]

print("Average petal width: ", petalWidthAverage)
print("Number of Iris-Versicolor flowers with petal width greater than the average: ", len(versicolorPetalWidthGreaterThanAverageeeeeeeeeeeeeeeeeeeeeeeeeee))
# print("Number of Iris-Versicolor flowers with petal width greater than the average: ", versicolorPetalWidthGreaterThanAverageeeeeeeeeeeeeeeeeeeeeeeeeee.shape[0]) #-----.shape[0]
```

### (11) Show count per name (number of rows for each species)

```python
countSetosa = df[df["species"] == "setosa"].shape[0]
countVersicolor = df[df["species"] == "versicolor"].shape[0]
countVirginica = df[df["species"] == "virginica"].shape[0]

print("Number of Iris-Setosa flowers: ", countSetosa)
print("Number of Iris-Versicolor flowers: ", countVersicolor)
print("Number of Iris-Virginica flowers: ", countVirginica)
```

Alternatively,

```python
countPerSpecies = df.groupby('species')['species'].count()
print("Count of flowers per species:\n\n", countPerSpecies)

# valueCounts = df.groupby('species')['species'].value_counts() #-----.value_counts()
# print(valueCounts)
```

### (12) Get the mean average per flower

```python
averagePerSpecies = df.groupby('species').mean()
print("Averages per flower species:\n\n", averagePerSpecies)
```
