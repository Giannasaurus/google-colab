## Using the `pandas` library

### (1) Import the packages

```python
import pandas as pd
```

### (2) Read the CSV file

```python
df = pd.read_csv('/content/drive/MyDrive/Colab-Notebooks/csv-files/iris.csv')
print(df)
```

### (3) Describe the iris dataset

```python
print(df.describe())
```

### (4) Display the first 10 rows

```python
print(df.head(10))
# print(df[0:10]) ----- alternative
```

### (5) Display the available columns

```python
print(df.columns.tolist())
# print(df.columns.values) ----- alternative: space-delimited list
```

### (6) Display the flowers with the name Iris-Virginica

```python
iris_virginica = df[df["species"] == "virginica"]
print(iris_virginica)
```

### (7) Display the flowers with the name Iris-Virginica or Iris-Versicolor

```python
iris_virginica_versicolor = df[(df["species"] == "virginica") | (df["species"] == "versicolor")]
print(iris_virginica_versicolor)
```

### (8) How many flowers have a petal length greater than 6.4?

```python
petal_length_gt_6pt4 = df[df["petal_length"] > 6.4].drop(["sepal_length", "sepal_width", "petal_width"], axis=1)

print(petal_length_gt_6pt4)
print("\nNumber of flowers with petal length greater than 6.4: ", len(petal_length_gt_6pt4))
# print("\nNumber of flowers with petal length greater than 6.4: ", petal_length_gt_6pt4.shape[0]) ----- alternative: .shape[0]
```

### (9) How many flowers have a petal length lower than 4?

```python
petal_length_lt_4 = df[df["petal_length"] < 4].drop(["sepal_length", "sepal_width", "petal_width"], axis=1)

print(petal_length_lt_4)
print("\nNumber of flowers with petal length lower than 4: ", len(petal_length_lt_4))
# print("\nNumber of flowers with petal length lower than 4: ", petal_length_lt_4.shape[0]) ----- alternative: .shape[0]
```

### (10) Among Iris-Versicolor, how many have a petal width greater than the average?

```python
petal_width_avg = df["petal_width"].mean()
versicolor_petal_width_gt_avg = df[(df["species"] == "versicolor") & (df["petal_width"] > petal_width_avg)]

print("Average petal width: ", petal_width_avg)
print("Number of Iris-Versicolor flowers with petal width greater than the average: ", len(versicolor_petal_width_gt_avg))
# print("Number of Iris-Versicolor flowers with petal width greater than the average: ", versicolor_petal_width_gt_avg.shape[0]) ----- alternative: .shape[0]
```

### (11) Show count per name (number of rows for each species)

```python
count_setosa = df[df["species"] == "setosa"].shape[0]
count_versicolor = df[df["species"] == "versicolor"].shape[0]
count_virginica = df[df["species"] == "virginica"].shape[0]

print("Number of Iris-Setosa flowers: ", count_setosa)
print("Number of Iris-Versicolor flowers: ", count_versicolor)
print("Number of Iris-Virginica flowers: ", count_virginica)
```

Alternatively,

```python
count_per_species = df.groupby('species')['species'].count()
print("Count of flowers per species:\n\n", count_per_species)
```

### (12) Get the mean average per flower

```python
avg_per_species = df.groupby('species').mean()
print("Averages per flower species:\n\n", avg_per_species)
```
